```python
import numpy as np
import pandas as pd
import cv2
import tensorflow as tf
from keras import Sequential
from keras import layers
from keras import models
```

```python
c_card=pd.read_csv("/content/drive/MyDrive/creditcard.csv")
```

```python
c_card.tail()
```

```python
c_card.isnull()
```

```python
np.random.seed(123)
data= np.random.rand(len(c_card))< 0.7
c_card_train=c_card[data]
c_card_test=c_card[~data]
```

```python
print(c_card_train.shape)
print(c_card_test.shape)
```

```python
c_card_tr_label=c_card_train.iloc[:,-1]
c_card_te_label=c_card_test.iloc[:,-1]
```

```python
print(c_card_tr_label.shape)
print(c_card_te_label.shape)
```

```python
c_card_tr_data=c_card_train.drop(columns="Class")
c_card_te_data=c_card_test.drop(columns="Class")
```

```python
mean=c_card_tr_data.mean()
c_card_tr_data -= mean
std=c_card_tr_data.std()
c_card_tr_data /= std
c_card_te_data -= mean
c_card_te_data /= std
```

```python
c_card_tr_data.dtypes
```

```python
network = models.Sequential()
network.add(layers.Dense(10, activation='relu', input_shape=(c_card_tr_data.shape[1],)))
network.add(layers.Dropout(0.5))
network.add(layers.Dense(8, activation='relu'))
network.add(layers.Dropout(0.5))
network.add(layers.Dense(6, activation='relu'))
network.add(layers.Dropout(0.5))
network.add(layers.Dense(1, activation="sigmoid"))
```

```python
network.compile(loss="binary_crossentropy",
                optimizer="Adam",
                metrics=["accuracy"])
```

```python
network.fit(c_card_tr_data, c_card_tr_label, epochs=100, validation_split=0.3)
```

```python
loss, accuracy = network.evaluate(c_card_te_data, c_card_te_label)
```

```python
print("Loss:", loss) 
print("Accuracy:", accuracy)
```
